##Importing Dataset

```python
emotions = load_dataset('SetFit/emotion')
```

```python
emotions
```

##Train,Test-Set Split

```python
def tokenize(data):
  return tokenizer(data['text'],padding=True,truncation=True)
```

```python
data=emotions.map(tokenize,batched=True,batch_size=None)
```

```python
data
```

```python
data.set_format('tf',columns=['input_ids', 'attention_mask', 'token_type_ids', 'label'])

BATCH_SIZE = 64

def group(inp):
    '''
    This function will group all the inputs of BERT
    into a single dictionary and then output it with
    labels.
    '''
    data = list(inp.values())
    return {
        'input_ids': data[1],
        'attention_mask': data[2],
        'token_type_ids': data[3]
    }, data[0]

train_dataset = tf.data.Dataset.from_tensor_slices(data['train'][:])
train_dataset = train_dataset.batch(BATCH_SIZE).shuffle(1000)
train_dataset = train_dataset.map(group, num_parallel_calls=tf.data.AUTOTUNE)

test_dataset = tf.data.Dataset.from_tensor_slices(data['test'][:])
test_dataset = test_dataset.batch(BATCH_SIZE)
test_dataset = test_dataset.map(group, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
inp, out = next(iter(train_dataset))
print(inp, '\n\n', out)
```

##Custom Bert Model with additional Fully-Connected Layer

```python
class BERTForClassification(tf.keras.Model):

    def __init__(self, bert_model, num_classes):
        super().__init__()
        self.bert = bert_model
        self.fc = tf.keras.layers.Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.bert(inputs)[1]
        return self.fc(x)
```

```python
classifier = BERTForClassification(model, num_classes=6)

classifier.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
history = classifier.fit(
    train_dataset,
    epochs=3
)
```

```python
classifier.evaluate(test_dataset)
```
